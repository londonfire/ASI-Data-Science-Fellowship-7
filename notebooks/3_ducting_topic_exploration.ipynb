# 3: Restaurant ducting fires

```python
import sys
sys.path.insert(0, '../')

import numpy as np
import pandas as pd
import pickle

from corpus_extractor import *

import data_cleaning
from data_cleaning import *

import topic_modellor
from topic_modellor import *

import mapping_gis
from mapping_gis import *


import matplotlib
import matplotlib.pyplot as plt

%pylab inline
pylab.rcParams['figure.figsize'] = (10, 10)

import seaborn as sns


import warnings
warnings.filterwarnings('ignore')
```

### Load data, corpus, and model

```python
lfb = pd.read_csv('../data/lfb_all_primary_fire_data.csv')
lfb = convert_date_time(lfb)
vector, vector_fit, model = pickle.load(open("../model/model_pickled.p", "rb"))
```

### Prepare topics data_frame

```python
lfb_free_text = CorpusExtractor().extract_free_text_only_data_frame(lfb)
```

```python
lfb_free_text['year'] = lfb_free_text['DDDateTimeOfCall'].dt.year
lfb_free_text['month'] = lfb_free_text['DDDateTimeOfCall'].dt.month
lfb_free_text['weekday'] = lfb_free_text['DDDateTimeOfCall'].dt.weekday
lfb_free_text['year-month'] = lfb_free_text['DDDateTimeOfCall'].map(lambda x: x.strftime('%Y-%m'))
```

### Ducting and restaurant ducting

```python
ducting = CorpusExtractor().extract_key_word_data_frame(lfb_free_text, 'ducting')
restaurant_ducting = ducting[ducting.ParentPropertyType=='Food and Drink']
```

### Compartmentation vs action based

```python
fig, axs = plt.subplots(2,1)
mapping = {'None': 0, 'Small': 1, 'Medium': 2, 'Large': 3, '5+': 4}

df = ducting
df['ActionBased'] = df.ActionBased.map(mapping)
ax = sns.boxplot(y="Compartmentation", x="ActionBased", data=df, ax=axs[0])

df = restaurant_ducting
df['ActionBased'] = df.ActionBased.map(mapping)
ax = sns.boxplot(y="Compartmentation", x="ActionBased", data=df, ax=axs[1])
```

### Ducting incidents per year

```python
fig, axs = plt.subplots(2,1)

df = ducting
ax = sns.countplot(x="year", data=df, color="black", ax=axs[0])
ax.set_ylim(0, 60)

df = restaurant_ducting
ax = sns.countplot(x="year", data=df, color="black", ax=axs[1])
ax.set_ylim(0, 60)
```

### Ducting incidents per month

```python
fig, axs = plt.subplots(2,1)

df = ducting
ax = sns.countplot(x="month", data=df, color="black", ax=axs[0])
ax.set_ylim(0, 60)

df = restaurant_ducting
ax = sns.countplot(x="month", data=df, color="black", ax=axs[1])
ax.set_ylim(0, 60)
```

### Ducting incidents per weekday

```python
fig, axs = plt.subplots(2,1)

df = ducting
ax = sns.countplot(x="weekday", data=df, color="black", ax=axs[0])
ax.set_ylim(0, 100)

df = restaurant_ducting
ax = sns.countplot(x="weekday", data=df, color="black", ax=axs[1])
ax.set_ylim(0, 100)
```

```python
df = restaurant_ducting

plt.figure(figsize=(15,7.5))
sns.set_context("notebook", font_scale=1.5, rc={"lines.linewidth": 2.5})
ax = sns.countplot(x="year", data=df, color="red")
ax.set_ylim(0, 40)
sns.set_style("whitegrid")
sns.plt.title('Fires in restaurants involving ducting systems', weight='bold', size=30)
ax.set_ylabel('')    
ax.set_xlabel('')
sns.despine(trim=True, offset=2)
plt.savefig("output.png")
```
