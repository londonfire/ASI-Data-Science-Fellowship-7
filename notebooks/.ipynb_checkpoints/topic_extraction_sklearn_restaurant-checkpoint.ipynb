# LDA and NMF using sklearn

```python
import pandas as pd
import numpy as np

import pyLDAvis
import pyLDAvis.sklearn
pyLDAvis.enable_notebook()

from sklearn.feature_extraction.text import TfidfVectorizer, CountVectorizer
from sklearn.decomposition import NMF, LatentDirichletAllocation

import sys
sys.path.insert(0, '../')

import free_text
from free_text import CorpusExtractor
from free_text import TopicModelling
```

```python
free_text_columns = ['BriefDescriptionOfFire','FurtherInformation']
lfb = pd.read_csv('../data/lfb_all_primary_fire_data.csv')
```

```python
n_topics = 5
n_top_words = 10
n_features = 2000
```

```python
lfb = lfb.loc[lfb['PropertyType'].isin(['Restaurant/cafe','Pub/wine bar/bar',"Takeaway, fast food"])]
```

```python
key_word_corpus = CorpusExtractor.extract_key_word_corpus(
    lfb, free_text_columns, 'ducting')

full_corpus = CorpusExtractor.extract_entire_corpus(lfb, free_text_columns)
```

```python
corpus = full_corpus
```

```python
TopicModelling.preprocess_corpus(corpus)
```

## NMF tf

```python
vector, vector_fit, model = TopicModelling.topic_modelling('NMF', 'tf', n_features, corpus, n_topics)
TopicModelling.print_top_words(vector, model, n_top_words)
```

## NMF tfidf

```python
vector, vector_fit, model = TopicModelling.topic_modelling('NMF', 'tfidf', n_features, corpus, n_topics)
TopicModelling.print_top_words(vector, model, n_top_words)
```

## LDA tf

```python
vector, vector_fit, model = TopicModelling.topic_modelling('LDA', 'tf', n_features, corpus, n_topics)
TopicModelling.print_top_words(vector, model, n_top_words)
```

```python
pyLDAvis.sklearn.prepare(model, vector_fit, vector)
```

```python
lda_tf = pyLDAvis.sklearn.prepare(model, vector_fit, vector)
pyLDAvis.save_html(lda_tfidf, '../topics/full_corpus_lda_tf.html')
```

## LDA tfidf

```python
vector, vector_fit, model = TopicModelling.topic_modelling('LDA', 'tfidf', n_features, corpus, n_topics)
TopicModelling.print_top_words(vector, model, n_top_words)
```

```python
pyLDAvis.sklearn.prepare(model, vector_fit, vector)
```

```python
lda_tfidf = pyLDAvis.sklearn.prepare(model, vector_fit, vector)
pyLDAvis.save_html(lda_tfidf, '../topics/full_corpus_lda_tfidf.html')
```
