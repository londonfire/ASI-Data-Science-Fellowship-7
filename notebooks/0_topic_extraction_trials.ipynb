## LDA TFIDF Classification

```python
topic_probability = model.transform(vector_fit)
topic_probability = np.delete(topic_probability, n_topics-1, axis=1) # remove topic 9 
topic_classified = pd.Series(np.argmax(topic_probability, 1))
```

```python
lfb_with_free_text = lfb.iloc[lfb[free_text_columns].dropna(how='all').index].reset_index()
lfb_with_free_text['topic_classified'] = (topic_classified +1)
```
